```python
# import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
#load data
data = pd.read_csv("https://raw.githubusercontent.com/KarianneKies/COVID-19-Temp/main/data/clean_data/data.csv")
```

```python
data.columns
```

```python
sample_data = data.sample(n=1000) #avoid overplotting
```

```python
#customize style 
custom_style = {'axes.axisbelow': True,
                'axes.edgecolor': 'white',
                'axes.facecolor': '#f0f0f0',
                'axes.grid': True,
                'axes.labelcolor': '.15',
                'axes.linewidth': 0,
                'font.family': 'Helvetica',
                'grid.color': 'white',
                'grid.linestyle': '-',
                'image.cmap': 'Greys',
                'legend.frameon': False,
                'legend.numpoints': 1,
                'legend.scatterpoints': 1,
                'lines.solid_capstyle': 'round',
                'pdf.fonttype': 42,
                'text.color': '.15',
                'xtick.color': '.15',
                'xtick.direction': 'out',
                'xtick.major.size': 0,
                'xtick.minor.size': 0,
                'ytick.color': '.15',
                'ytick.direction': 'out',
                'ytick.major.size': 0,
                'ytick.minor.size': 0}


sns.set_style(rc=custom_style)


graph = sns.lmplot(x = "TAVG", y = "new_cases", data = sample_data, height=8, aspect=1.2,  
                   line_kws={'color': '#FFA500'}, scatter_kws={'alpha':0.4,'color':'#808080'})

ax = plt.gca() #get axes 
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)

# Customize the axes
#ax.axhline(y = 0, color = 'black', linewidth = 2, alpha = .3)

ax.set_xlabel("Average Temperature", fontsize = 19, alpha = .85)
ax.set_ylabel("New Cases Smoothed", fontsize = 19, alpha = .85)

# Customize title 
ax.text(x = -10.5, y =62000, s = "The effect of temperature on COVID-19 infection rate",
               fontsize = 26, weight = 'bold', alpha = .85)
ax.text(x = -10.5, y = 56500,
              s = 'The average temperature in C plotted against a moving average from the day before \nto the day after each date shown.',
              fontsize = 19, alpha = .5)



ax.text(x = -10.5, y = -12000,
    s = '                                                                                                                                                Sources: OWID & NOAA',fontsize = 14, color = 'grey', alpha=.7)

plt.savefig('/Users/kariannekies/Documents/COVID-Temp/lmplotCOVID19_Temp.png')
plt.show()


```
